```python
import matplotlib.pyplot as plt
```

```python
# What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("X Data")
plt.ylabel("Y Data")
plt.title("Random Data Graph")
```

```python
#Why is the subplot() function used? Draw four line plots using the subplot() function
# It is used to add plot to our figure.
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
fig=plt.figure(figsize=(10,8))
plot1=fig.add_subplot(2,2,1)
plot1.plot(x1,y1,c="red")
plt.title("Plot 1")
plot2=fig.add_subplot(2,2,2)
plot2.plot(x2,y2,c="blue")
plt.title("Plot 2")
plot3=fig.add_subplot(2,2,3)
plot3.plot(x3,y3,c="green")
plt.title("Plot 3")
plot4=fig.add_subplot(2,2,4)
plot4.plot(x4,y4,c="yellow")
plt.title("Plot 4")
plt.show()
```

```python
# What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 
# A bar plot, also known as a bar chart or bar graph, is a type of graph that represents categorical data with rectangular bars. 
# The length or height of each bar is proportional to the frequency of data points in each category. 
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit,color="red")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Bar Plot")
```

```python
plt.barh(company,profit,color="blue")
plt.ylabel("Company")
plt.xlabel("Profit")
plt.title("Horizonatal Bar Plot")
```

```python
# What is a box plot? Why is it used? Using the following data plot a box plot.
# A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset along a single axis.
# It summarizes key features of the dataset such as its central tendency, dispersion, and skewness. 
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
```
