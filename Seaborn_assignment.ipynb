```python
import seaborn as sb
```

```python
# Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot
# relplot-It is use to plot the graph when we have to analyze the data in a relational manner ,means if we want to identify what
# what the other attributes are behaving with respect to x and y.
# barplot -Used to generate bargraph.
# histplot and displot - Used to generate histogram.
# scatter plot- Used to generate a pointed graph. 
# joint plot- Generates a graph that contains both scatter and histplot
# pair plot- Generates multiplegraphs in one graph for each and every possible pairs in data.
```

```python
tips=sb.load_dataset("tips")
```

```python
tips.head(3)
```

```python
sb.relplot(x=tips["total_bill"],y=tips["tip"],hue="smoker",style="size",data=tips)
```

```python
# Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" 
#for different events and regions. 
fmri=sb.load_dataset("fmri")
```

```python
fmri.head(3)
```

```python
sb.lineplot(x=fmri["timepoint"],y=fmri["signal"],hue="event",style="region",data=fmri)
```

```python
# Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using 
#x = 'pclass', y = 'age' and y = 'fare'.
titan=sb.load_dataset("titanic")
```

```python
sb.boxplot(x=titan["pclass"],y=titan.age)
```

```python
sb.boxplot(x=titan["pclass"],y=titan["fare"])
```

```python
#  Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue 
# parameter for the 'cut' column of the diamonds dataset. 
dia=sb.load_dataset("diamonds")
```

```python
sb.displot(x=dia["price"],hue=dia["cut"])
```

```python
#Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset. 
iris=sb.load_dataset("iris")
```

```python
try:
    sb.pairplot(data=iris,hue=iris["species"])
except ValueError as e:
    print(e)
    sb.pairplot(data=iris)
```
